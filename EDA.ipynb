```python
from hda.eda import load_sample_ecg
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20,5]
plt.style.use('ggplot')
```

```python
#      Experiment Subject  Date States 
TEST_FILE = 'HaLT-SubjectA-160223-6St-LRHandLegTongue.mat'
DATA_FOLDER = '../data/'
```

```python
mat = load_sample_ecg(TEST_FILE, DATA_FOLDER, as_pandas=False)
df = load_sample_ecg(TEST_FILE, DATA_FOLDER, as_pandas=True)
```

```python
# The eGUI [interaction record of the recording session]

#  0 - Blank
#  1 - Left Hand
#  2 - Right Hand
#  3 - Passive / Neutral
#  4 - Left Leg
#  5 - Tongue
#  6 - Right Leg
# 91 - Session Break
# 92 - Experiment End
# 99 - Initial Relaxation

display(mat['o']['marker'][0][0][100000:100005])
extracted_data = mat['o']['marker'][0][0].flatten()

# Plotting
plt.plot(extracted_data)
plt.title("Data Plot")
plt.xlabel("Sample Index")
plt.ylabel("Value")
plt.show()
```
